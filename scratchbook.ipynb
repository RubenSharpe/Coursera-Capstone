```python
path = '~/Documents/Projects/Coursera-Capstone/Neighbourhoods.csv'
nbh_data = pd.read_csv(path)
nbh_data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
nbh_data.head()
```

```python
# Separate combined neighbourhoods
apple = nbh_data.assign(Neighbourhood=nbh_data['Neighbourhood'].str.split(',')).explode('Neighbourhood')
apple.head()
```

```python
apple = apple.groupby(['Neighbourhood']).mean()
apple = apple.sort_values(by='Neighbourhood')
```
